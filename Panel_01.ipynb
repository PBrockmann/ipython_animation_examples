```python
%%ferret -q -s 600,800

use "http://esgf.extra.cea.fr/thredds/dodsC/store/p86broc/ANIM/v3.historical1_19500101_20051231_1D_t2m.nc"
use "http://esgf.extra.cea.fr/thredds/dodsC/store/p86broc/ANIM/v3.historical1_18500101_20051231_1D_sosstsst.nc"
def sym datestr=15-JUL-2000
let temp=t2m[d=1]-273.15
def sym qualifiers=/nolab/lev=(-inf)(-20,30,2)(inf)/pal=fast_001
set v upper ; go margins_set 10 10 10 10 ; shade($qualifiers) temp[t=($datestr)] ; go land
set v lower ; go margins_set 10 10 10 10 ; shade/modulo($qualifiers) sosstsst[t=($datestr)], nav_lon, nav_lat ; go land
go text_put 50 98 "($datestr)" 0 0.5
```

```python
import IPython.display
import jinja2
import glob

def display_animation(animID, files, interval=200, style=False):
    env = jinja2.Environment(loader=jinja2.FileSystemLoader('.'))
    template = env.get_template('animation_template.html')

    try:            # frames are local and can be expanded
        listFrames = sorted(glob.glob(files))
    except:         # frames are remote ressources
        listFrames = files
    htmlSrc = template.render(ID=animID, listFrames=listFrames, interval=interval, style=style)
    
    return IPython.display.HTML(htmlSrc)
```

```python
import os, shutil
shutil.rmtree("frames1", ignore_errors=True)
os.mkdir("frames1")
i=1
#for month in ["JAN","FEB"]:
for month in ["JAN","FEB","MAR","APR","MAY","JUN","JUL","AUG","SEP","OCT","NOV","DEC"]:
    %ferret_run -q -s 20,20 \
        'def sym datestr=15-%(month)s-2000 ;\
        set v upper ; go margins_set 10 10 10 10 ; shade($qualifiers) temp[t=($datestr)] ; go land ;\
        set v lower ; go margins_set 10 10 10 10 ; shade/modulo($qualifiers) sosstsst[t=($datestr)], nav_lon, nav_lat ; go land ;\
        frame/file="frames1/frame_%(i)03d.png"/xpixels=500 \
        ' %locals()
    iStr=str(i).zfill(3)
    monthStr=str(i).zfill(2)
    #!echo {month} {istr}
    !convert -font FreeMono-Bold -fill black -gravity Center -density 150 -draw 'text +0,+5 "2000/{monthStr}"' frames1/frame_{iStr}.png frames1/tmp.png  
    !mv frames1/tmp.png frames1/frame_{iStr}.png
    i=i+1
```
